# Load Sample Data and Explore Data Structure

```python
import pandas as pd

# Direct download from HuggingFace
url = "https://huggingface.co/datasets/eloukas/edgar-corpus/resolve/refs%2Fconvert%2Fparquet/year_1993/validation/0000.parquet"
df = pd.read_parquet(url)

print(f"Loaded {len(df)} documents")
print(df.columns)
```
